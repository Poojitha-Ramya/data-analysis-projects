```python
import numpy as np
```

```python
Student_DataBase={
    "Name":["Alice","John","Ramya","Reshu", "Sam", "Emily", "David", "Sarah", "Michael", "Jessica", "Christopher", "Ashley", "Matthew", "Amanda", "Joshua", "Brittany", "Andrew", "Lauren", "Daniel", "Megan", "James", "Hannah", "Joseph", "Nicole", "Ryan", "Alexis", "Nicholas", "Samantha", "William", "Elizabeth", "Jonathan", "Stephanie", "Tyler", "Tiffany", "Brandon", "Melissa", "Justin", "Jennifer", "Zachary", "Kimberly"],
    "Maths":np.random.randint(35,100,40),
    "Science":np.random.randint(30,100,40),
    "English":np.random.randint(25,100,40),
    "Chemistry":np.random.randint(40,100,40),
    "Physics":np.random.randint(29,100,40)
}
```

```python
import pandas as pd
dfstudents=pd.DataFrame(Student_DataBase)
```

```python
dfstudents.to_csv('Student_data.csv',index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Student_data.csv")
print(df)
```

```python
df["Average"]=df[["Maths","Science","English","Chemistry","Physics"]].mean(axis=1)
print(df)
```

```python
Subject_average=df[["Maths","Science","English","Chemistry","Physics"]].mean()
print(Subject_average)
```

```python
plt.bar(Subject_average.index,Subject_average.values)
plt.xlabel("Subject")
plt.ylabel("Average Marks")
plt.title("Subject Performance")
plt.show()
```

```python
plt.scatter(df['English'],df['Science'],alpha=0.4,edgecolors='red')
plt.xlabel("English Marks")
plt.ylabel("Science Marks")
plt.title("English vs Science Marks")
plt.show()
```

```python
plt.figure(figsize=(40,20))
plt.bar(df["Name"],df["Average"])
plt.xlabel("Subject")
plt.ylabel("Average Marks")
plt.title("Subject Performance")
plt.show()
```

```python
topper=df.loc[df['Average'].idxmax()]
print("topper:\n",topper)
```

```python
fail_students=df.loc[(df[["Maths","Science","English","Chemistry","Physics"]] < 35).any(axis=1)]
print("Failing Students:\n", fail_students)
```

```python
top_students=df.loc[df['Average'] > 80, ['Name', 'Average']]
print("Top Students:\n", top_students)
```

```python
import matplotlib.pyplot as plt
toppers=len(top_students)
loosers=len(fail_students)
sizes=[toppers,loosers]
Students=['top_students','fail_students']
plt.pie(sizes,labels=Students,autopct='%1.1f%%',colors=['pink','skyblue'])
plt.show()
```
